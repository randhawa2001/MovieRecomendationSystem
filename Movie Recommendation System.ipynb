```python
from sklearn.metrics.pairwise import cosine_similarity
selected_columns = ['Year', 'Length', 'Popularity','Awards','id']
selected_data = df[selected_columns]
cosine_sim = cosine_similarity(selected_data)
```

```python
def get_similar_movies(movie_index, similarity_matrix, top_n=10):
    similarity_scores = list(enumerate(similarity_matrix[movie_index]))
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
    similar_movies = [movie[0] for movie in similarity_scores[1:top_n+1]]
    return similar_movies
```

```python
movie_index = 0 
similar_movies = get_similar_movies(movie_index, cosine_sim, top_n=10)
print("Top 10 recommended movies:")
for movie_index in similar_movies:
    print(df.iloc[movie_index])
```
